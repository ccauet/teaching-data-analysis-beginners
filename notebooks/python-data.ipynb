```python
from sklearn import datasets

import pandas as pd

import matplotlib
import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
iris['target_names']
```

```python
df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
```

```python
df = df.assign(target=iris['target'])
```
